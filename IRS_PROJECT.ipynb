# **Read A Text File**

```python
documents_= open("/content/drive/MyDrive/Colab Notebooks/IRSdataset.txt","r",encoding='ISO-8859-1')

documents=documents_.read()
document = documents.split('\n')
print(document[0:2])
```

# **Remove NewLines ( '\n' )**

```python
import re
lexicon_without_newlines=[]
for doc in document:
    lexicon_without_newlines.append(re.sub(r'\\n', ' ', doc))
print(lexicon_without_newlines[0:2])
```

# **Tokenization**

```python
import nltk
#!pip install nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
tokens=[]
# Tokenize the document into words
for i in lexicon_without_newlines:
    tokens.append(word_tokenize(str(i)))
print(tokens[0:2])
```

# **Remove Punctuation Marks**

```python
import string

# Define a translation table to remove punctuation
translator = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
lexicon_without_punctuation=[]
# Remove tokens containing punctuation marks
for i in tokens:
    lexicon_without_punctuation.append([word.translate(translator) for word in i if word.translate(translator).strip()])

print(lexicon_without_punctuation[0])
```

# **Remove Digits & Characters**

```python
# Remove tokens containing digits
lexicon_without_digits=[]
for i in lexicon_without_punctuation:
    lexicon_without_digits.append([word for word in i if ((word.isalpha()) and (len(word) > 1))])

print(lexicon_without_digits[0])
```

# **Remove Stop Word**

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

# Get English stopwords from NLTK
stop_words = set(stopwords.words('english'))

lexicon_without_stopword=[]
# Remove stopwords
for i in lexicon_without_digits:
    lexicon_without_stopword.append([word for word in i if word.lower() not in stop_words])
print(lexicon_without_stopword[0])
```

# **Stemming**

```python
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer
stemmer = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")
stemmed_tokens_porter=[]
stemmed_tokens_lancaster=[]
stemmed_tokens_snowball=[]
# Apply stemming to each word in the sentence
for i in lexicon_without_stopword:
  stemmed_tokens_porter.append([stemmer.stem(word) for word in i])
  stemmed_tokens_lancaster.append([lancaster.stem(word) for word in i])
  stemmed_tokens_snowball.append([snowball.stem(word) for word in i])
#extend words from list of lists
words_porter=[]
words_lancaster=[]
words_snowball=[]
for lists in stemmed_tokens_porter:
  for i in lists:
    if len(i)>1:
      words_porter.append(i)
for lists in stemmed_tokens_lancaster:
  for i in lists:
    if len(i)>1:
      words_lancaster.append(i)
for lists in stemmed_tokens_snowball:
  for i in lists:
    if len(i)>1:
      words_snowball.append(i)
```

```python
print('text before lemmatization \nWords : ',lexicon_without_stopword[0:2])
print('')
print('Total of stemmed words using Poter Stemmer: ',len(words_porter))
print('Stemmed words : ',words_porter[0:100])
print('')
print('Total of stemmed words using lancaster: ',len(words_lancaster))
print('Stemmed words : ',words_lancaster[0:100])
print('')
print('Total of stemmed words using Snowball: ',len(words_snowball))
print('Stemmed words : ',words_snowball[0:100])
```

# **Lemmatization**

```python
!pip install pattern
```

```python
from nltk.stem import WordNetLemmatizer
from pattern.en import lemma
from textblob import Word
#lemmatization
lemmatized_tokens_wordnetlemmatizer=[]
lemmatized_tokens_pattern=[]
lemmatized_tokens_textblob=[]
for i in lexicon_without_stopword:
  lemmatized_tokens_wordnetlemmatizer.append([WordNetLemmatizer().lemmatize(word) for word in i])
  lemmatized_tokens_pattern.append([lemma(word) for word in i])
  lemmatized_tokens_textblob.append([Word(word).lemmatize() for word in i])
#case folding
casefolded_wordnetlemmatizer=[]
casefolded_pattern=[]
casefolded_textblob=[]
for i in lemmatized_tokens_wordnetlemmatizer:
  casefolded_wordnetlemmatizer.append([string.casefold() for string in i])
for i in lemmatized_tokens_pattern:
  casefolded_pattern.append([string.casefold() for string in i])
for i in lemmatized_tokens_textblob:
  casefolded_textblob.append([string.casefold() for string in i])
#extend words from list of lists
words_wordnetlemmatizer=[]
words_pattern=[]
words_textblob=[]
for lists in casefolded_wordnetlemmatizer:
  for i in lists:
    if len(i)>1:
      words_wordnetlemmatizer.append(i)
for lists in casefolded_pattern:
  for i in lists:
    if len(i)>1:
      words_pattern.append(i)
for lists in casefolded_textblob:
  for i in lists:
    if len(i)>1:
      words_textblob.append(i)
```

```python
print('text before lemmatization \nWords : ',lexicon_without_stopword[0:2])
print('')
print('Total of Words using WordNetLemmatizer : ',len(words_wordnetlemmatizer))
print('Words : ',words_wordnetlemmatizer[0:100])
print('')
print('Total of Words using  pattern: ',len(words_pattern))
print('Words : ',words_pattern[0:100])
print('')
print('Total of Words using textblob : ',len(words_textblob))
print('Words : ',words_textblob[0:100])
```

```python
#choose pattern function to get words from

#remove doublecate
words=list(set(words_pattern))
#sorting
words.sort()
#covert from list of lists to list of strings
list_of_strings = [' '.join(inner_list) for inner_list in casefolded_pattern]
print(len(words))
```

# **Bag-Of-Words**

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
# Convert vocabulary that produces from pattern into a bag-of-words representation

vectorizer = CountVectorizer(vocabulary=words)
matrix = vectorizer.fit_transform(list_of_strings)
bag_of_words=pd.DataFrame(matrix.toarray(),columns=words)
```

```python
# Display the bag-of-words representation
print("Bag-of-Words Representation:")
bag_of_words
```

# **TF-IDF Matrix**

```python
from sklearn.feature_extraction.text import TfidfTransformer

# Convert the bag-of-words matrix to TF-IDF representation using TfidfTransformer
tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(matrix)

# Convert the TF-IDF matrix to a DataFrame for visualization
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=words)
```

```python
# Print the TF-IDF representation
print("TF-IDF feature space:")
tfidf_df
```

# **Search For Top 10 Document Match Your Query**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Assuming 'documents' is a list of documents
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(list_of_strings)

# Vectorize your query/document
query = input('write your query here : ')
query_vector = tfidf_vectorizer.transform([query])

# Calculate cosine similarity between query and all documents
similarities = cosine_similarity(query_vector, tfidf_matrix)

# Get indices of documents sorted by similarity to the query
similar_documents_indices = similarities.argsort()[0][::-1]

# Retrieve top N similar documents
top_n = 10
top_similar_documents = [document[i] for i in similar_documents_indices[:top_n]]
for i in range(10):
    print('Document',i+1,'-',top_similar_documents[i])
```

```python
from sklearn.model_selection import cross_val_score, KFold, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

# Load the iris dataset
data = load_iris()
X, y = data.data, data.target

# Outer loop for model selection
outer_cv = KFold(n_splits=5, shuffle=True, random_state=42)
print(outer_cv)
# Inner loop for hyperparameter tuning
inner_cv = KFold(n_splits=3, shuffle=True, random_state=42)
print(inner_cv)
# Define the model (Random Forest in this case)
model = RandomForestClassifier()

# Define hyperparameters to search
param_grid = {'n_estimators': [50, 100, 150], 'max_depth': [None, 10, 20]}

# Nested cross-validation
nested_cv = GridSearchCV(estimator=model, param_grid=param_grid, cv=inner_cv)
nested_score = cross_val_score(nested_cv, X=X, y=y, cv=outer_cv)

# Output the nested cross-validation results
print("Nested Cross-Validation Scores: ", nested_score)
print("Average Nested Cross-Validation Score: {:.2f}".format(nested_score.mean()))
```
